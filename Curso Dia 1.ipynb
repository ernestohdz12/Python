## Operaciones matematicas


```python
# 4to nivel de importancia
# Suma ordinaria
print ("Suma Ordinaria")
print (3 + 4) # un 7
variable1,variable = 20,40 #60
print(variable1 + variable)

# Resta ordinaria
print ("Resta ordinaria")
print (2 - 5) # -3

#Declaracion de variables
v1 = 10.4
v2 = 15
resultado = v2 - v1
print(resultado) # 4.6
```

```python
# 3er nivel de importancia
#Multiplicacion ordinaria
print ("Multiplicacion ordinaria")
M1, M2 = 2,3.14
print (M1 * M2) # 6.28
print ("-------")
# Division ordinaria
print ("Division ordinaria")
M1,M2,M3,M4,M5,M6 = 2,5,6,5,4,5
print(M1 / M2)
print(M3 / M4)
print(M5 / M6)
print ("------")
# Division de piso
print (M1 // M2)
print (M3 // M4)
print (M5 // M6)
```

```python
print ("-----")
# Potencias
print (2 ** 3) # 8
P1,P2 = 7,2
print( P1 ** P2) # 49

```
